```python
import pandas as pd

path = r"C:\Users\Gabi Fuller\Documents\PROGRAMMING MACHINE LEARNING A - 4152024 - 1128 AM\adult-modified.csv"
data = pd.read_csv(path, sep = ',')

df = pd.DataFrame(data)

print('data overview')
print(df.info())

print("\nDescriptive statistics for numeric attributes:")
print(df.describe(include='all'))

print("\nFrequency counts for categorical attributes:")
for col in df.select_dtypes(include=['object']):
  print(df[col].value_counts())
```

```python
#2
import matplotlib.pyplot as plt
import numpy as np

#boxplot
numeric_columns = ['age', 'hours-per-week', 'education']
plt.figure(figsize=(10, 6))
plt.boxplot(df[numeric_columns], vert=False, patch_artist=True, notch=True, medianprops={"linewidth": 2})
plt.xlabel('Values')
plt.ylabel('Variables')
plt.title('Box Plot of Numeric Variables')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#histogram

hist_columns = ['age', 'hours-per-week', 'education']
plt.figure(figsize=(12,6))

for i, col in enumerate(hist_columns):
    plt.subplot(1, 3, i+1)
    plt.hist(df[col])
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {col}')

plt.tight_layout()

# scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['hours-per-week'])
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.title('Scatter Plot of Age vs. Hours per Week')
plt.grid(True)
plt.show()
```

```python
#3

workclass = df['workclass'].value_counts()
marital_status = df['marital-status'].value_counts()
race = df['race'].value_counts()
sex = df['sex'].value_counts()
income = df['income'].value_counts()

#bar charts
#marital status
plt.bar(marital_status.index, marital_status.values)
plt.title('Bar Chart of Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Frequency')
plt.show()

#workclass
plt.bar(workclass.index, workclass.values)
plt.title('Bar Chart of Work-Class')
plt.xlabel('Work-Class')
plt.ylabel('Frequency')

plt.show()

#race
plt.bar(race.index, race.values)
plt.title('Bar Chart of Race')
plt.xlabel('Race')
plt.ylabel('Frequency')

plt.show()

#sex
plt.bar(sex.index, sex.values)
plt.title('Bar Chart of Sex')
plt.xlabel('Sex')
plt.ylabel('Frequency')

plt.show()

#income
plt.bar(income.index, income.values)
plt.title('Bar Chart of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')

plt.show

# Display the bar chart
plt.show()
```

```python
#4 crosstab/barcharat


workclass_income_table = pd.crosstab(df['workclass'], df['income'])

# workclass vs income 
plt.figure(figsize=(10, 6))
workclass_income_table.plot(kind='bar', stacked=False)
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Income Distribution Workclass')
plt.xticks(rotation=45)
plt.legend(title='Income')
plt.tight_layout()
plt.show()

# Cross-tab of race vs income (focusing on <=50K category)
race_income_table = pd.crosstab(df['race'], df['income'])
race_percentages = (race_income_table['<=50K'] / race_income_table.sum(axis=1)) * 100

# race percentages in <=50K category vs income 
plt.figure(figsize=(8, 6))
race_percentages.plot(kind='bar')
plt.xlabel('Race')
plt.ylabel('Percentage in <=50K')
plt.title('Percentage in <=50K Income vs Race')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Observations:
# From the workclass vs income chart, certain workclasses have a higher number of individuals in the <=50K income category.
# The race percentages in <=50K income chart highlights how much of an income disparity there is of white vs non-white individuals
```

```python
#5
low_inc = df[df['income'] <= '<=50K']
high_inc = df[df['income'] > '<=50K']

summary = pd.DataFrame({
    'Variable': ['age', 'education', 'hours-per-week'],
    'Low Income': [low_inc[variable].mean() for variable in ['age', 'education','hours-per-week']],
    'High Income': [high_inc[variable].mean() for variable in ['age', 'education','hours-per-week']]    
})

#bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(summary['Variable'], summary['Low Income'], label='Low Income')
ax.bar(summary['Variable'], summary['High Income'], bottom=summary['Low Income'], label='High Income')
ax.set_title('low vs high income')
ax.set_xlabel('Variable')
ax.set_ylabel('Mean value')
ax.legend()
plt.show()
```
